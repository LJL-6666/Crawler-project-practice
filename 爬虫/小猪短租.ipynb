```python
import os
os.getcwd()
os.chdir("G:/CDO/数据存储和获取")
import re
import requests
import re
import pandas as pd
```

```python
response = requests.get('http://bj.xiaozhu.com/')
res=response.text
print(res)
```

```python
#爬取标题title
title=re.compile('<li\slodgeunitid.*?title=(.*?)\sdata',re.S)
title=re.findall(title,res)
title
```

```python
#价格
price=re.compile('<span\sclass="result_price".*?i>(.*?)</i>(.*?)</span>',re.S)
price=re.findall(price,res)
price_n=[]
for item in price:
    price_n.append(item[0]+item[1])
price_n
```

```python
#房型
pattern_h=re.compile('<em\sclass="hiddenTxt">(.*?)<',re.S)
hiddenTxt=re.findall(pattern_h,res)
hiddenTxt_new=[h.strip() for h in hiddenTxt]
hiddenTxt_new
```

```python
#评论
com=re.compile('<span\sclass="commenthref">(.*?)</span>',re.S)
comment=re.findall(com,res)
com=[i.strip() for i in comment]
com
```

```python
columns={'标题':title,'价格':price_n,'出租形式':hiddenTxt_new,'点评':com}
data_one=pd.DataFrame(columns)
data_one.iloc[:7,:]
```

```python
data_one_1=data_one.iloc[data_one['点评'][data_one['点评'].str.contains(r'分')].index,]
data_one_1
```

```python
data_one_2=data_one.drop(list(data_one['点评'][data_one['点评'].str.contains(r'分')].index))
data_one_2
```

```python
data_one_2['点评new']=['    /']*len(data_one_2)
data_one_2
```

```python
data_one_2['点评new']=data_one_2['点评new']+data_one_2['点评']
data_one_2
```

```python
data_one_2_new=data_one_2.drop('点评',axis=1)
data_one_2_new
```

```python
data_one_2_finall=data_one_2_new.rename(columns={'点评new':'点评'})
data_one_2_finall
```

```python
data_f=pd.concat([data_one_1,data_one_2_finall])
data_f.iloc[:7,:]
```

```python
urllist=[]
for i in range(10):
    urllist.append('http://bj.xiaozhu.com/search-duanzufang-p'+str(i)+'-0/')
urllist
```

```python
data = pd.DataFrame(columns={'标题','价格','出租形式','点评'})
data
```

```python
for url in urllist:
    response = requests.get(url)
    res = response.text
    
    title=re.compile('<li\slodgeunitid.*?title=(.*?)\sdata',re.S)
    title=re.findall(title,res)
    
    price=re.compile('<span\sclass="result_price".*?i>(.*?)</i>(.*?)</span>',re.S)
    price=re.findall(price,res)
    price_n=[]
    for item in price:
        price_n.append(item[0]+item[1])
    
    pattern_h=re.compile('<em\sclass="hiddenTxt">(.*?)<',re.S)
    hiddenTxt=re.findall(pattern_h,res)
    hiddenTxt_new=[h.strip() for h in hiddenTxt]
    
    com=re.compile('<span\sclass="commenthref">(.*?)</span>',re.S)
    comment=re.findall(com,res)
    com=[i.strip() for i in comment]

    columns={'标题':title,'价格':price_n,'出租形式':hiddenTxt_new,'点评':com}
    data_o=pd.DataFrame(columns)
    data=pd.concat([data,data_o])
    
```
