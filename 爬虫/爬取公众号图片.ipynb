```python
import requests    
import os      
os.getcwd()
os.chdir("G:/CDO/数据存储和获取/爬公众号图片/")
from lxml import etree
```

```python
def start():
#第一层，获取公众号全部文章的链接
    filename='1602.tmp.txt'
    with open(filename,'r',encoding='utf-8')as f:
        source=f.read()  
    html_ele=etree.HTML(source) 
    hrefs=html_ele.xpath('//div[contains(@data-type,"APPMSG")]/h4/@hrefs')      
    num=0
    for i in hrefs:
        num+=1
        try:
            apply_one(i)
        except:
            continue
        print('第%d篇爬取完毕'%num)
```

```python
#第二层，解析单篇文章
def apply_one(url):
    headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36'
        }
    response=requests.get(url,headers=headers)
    elements=etree.HTML(response.text)
    data_src=elements.xpath('//section[contains(@style,"text-align")]/section/img/@data-src')
    data_src=data_src[2:-1]
    for src in data_src:
        try:
            download(src)  
        except:
            continue
```

```python
#第三层下载层
def download(src):
    headers={
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36'}
    response=requests.get(src,headers=headers)
    name=src.split('/')[-2]  #截取文件名
    dtype=src.split('=')[-1]  #截取图片类型
    name+='.'+dtype     #重构图片名
    os.makedirs('G:/CDO/数据存储和获取/爬公众号图片/doge',exist_ok=True) 
    with open('G:/CDO/数据存储和获取/爬公众号图片/doge'+name,'wb')as f:
        f.write(response.content)
```

```python
if __name__=='__main__':
    start()
```
