```python
def num_change(word_cmap,number):
    list=number.replace('&#','').split(';')
    list.pop(-1)
    word_count=''
    for va in list:
        word_count+=word_cmap[int(va)]
    return(word_count)
```

```python
num_finall=[]
for number in numberlist:
    num_finall.append(num_change(word_cmap,number))
```
