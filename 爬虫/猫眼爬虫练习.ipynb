```python
s1=[]
s2=[]
s3=[]
s4=[]
s5=[]
for i in range(10):
    res=requests.get('https://maoyan.com/board/4?offset='+str(i*10),headers={'User-Agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Mobile Safari/537.36'})
    res.encoding='utf8'
    print(res.request.url)
    soup=BeautifulSoup(res.text,'lxml')
    s1=s1+[n.text for n in soup.find_all(name='p',class_='name')]
    s2=s2+[re.search("(.*?)@",v).group() for v in [m.attrs['data-src'] for m in soup.find_all(name='img',class_="board-img")]]
    s3=s3+[re.sub('主演：','',r) for r in [re.sub('\n|\s+','',g) for g in[a.text for a in soup.find_all(name='p',class_='star')]]]
    s4=s4+[re.sub('\(.*?\)','',k) for k in [re.sub('上映时间：','',l) for l in [t.text for t in soup.find_all(name='p',class_="releasetime")]]]
    s5=s5+[s.text for s in soup.find_all(name='p',class_='score')]
```

```python
headers = {'User-Agent':"Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Mobile Safari/537.36'"}

def getpage(url):
    response = requests.get(url)
    if response.status_code == 200:
        data= response.text
        return data
    else:
        return None

def getinfo(html):
    soup = BeautifulSoup(html,'lxml')   
    items = soup.select('dd')           

    for item in items:
        index = item.find(name = 'i',class_='board-index').get_text()
        name = item.find(name = 'p',class_='name').get_text()
        start = item.find(name = 'p',class_='star').get_text().strip()
        time = item.find(name = 'p',class_='releasetime').get_text()
        score = item.find(name = 'p',class_ = 'score').get_text()

        yield {
            'index':index,
            'name':name,
            'star':start,
            'time':time,
            'score':score
        }


def writeData(field):
    with open('G:/CDO/数据存储和获取/maoyan_info.txt','a',encoding='utf-8')as f:
        f.write(json.dumps(field,ensure_ascii=False) + '\n')


if __name__ == '__main__':
    for i in range(10):
        url="https://maoyan.com/board/4?offset="+str(10*i)
        html = getpage(url)
        for item in getinfo(html):
            print(item)
            writeData(item)
```
