```python
import os
os.getcwd()
os.chdir("G:/CDO/数据存储和获取/")
import requests
import re
import xlwt
import pandas as pd
from bs4 import BeautifulSoup
from lxml import etree
```

```python
def get_html(url):
    import requests

    html = requests.get(url)
    return html


url = 'https://maoyan.com/board/4'
html = get_html(url)
html
```

```python
web_text = html.text
print(web_text)
```

```python
html = etree.HTML(web_text)#将xml的文件转化为HTML
result = etree.tostring(html)
print(result.decode('utf-8'))
with open('result.html','w+',encoding='utf-8') as f:
    f.write(result.decode())
```

```python
Html = etree.parse('./result.html', etree.HTMLParser())
result = etree.tostring(Html)
print(result)
```
